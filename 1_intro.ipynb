# IRIS CLASSIFICATION

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print("Keys iris_dataset: \n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset['DESCR'][:193] + "\n..")
```

```python
print("target name: {}".format(iris_dataset['target_names']))
```

```python
print("feature names: \n{}".format(iris_dataset['feature_names']))
```

```python
print("Data array type: {}".format(type(iris_dataset['data'])))
```

```python
print("Data array format: {}".format(iris_dataset['data'].shape))
```

```python
print("First five array examples:\n{}".format(iris_dataset['data'][:5]))
```

```python
print("Type target array:\n{}".format(type(iris_dataset['target'])))
```

```python
print("Target array shape: {}".format(iris_dataset['target'].shape))
```

```python
print("Target:\n{}".format(iris_dataset['target']))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], 
                                                    iris_dataset['target'], random_state = 0 )
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```python
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
grr = pd.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15), marker='o',
                       hist_kwds = {'bins':20}, s=60, alpha = .8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5,2.9,1,0.2]])
print("array shape X_new: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target: {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Prediction for test dataset:\n {}".format(y_pred))
```

```python
print("Correctness in test data: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Correctness in test data: {:.2f}".format(knn.score(X_test,y_test)))
```
